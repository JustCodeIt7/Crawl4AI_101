```python
import asyncio
import time
from typing import Dict, Any
import requests
from bs4 import BeautifulSoup
from crawl4ai import AsyncWebCrawler

# Test URL for our comparison
TEST_URL = "https://www.python.org"
```

## 1. BeautifulSoup - The Traditional Approach

### Pros:

- Simple to learn
- Good for basic HTML parsing

### Cons:

- ❌ **No markdown output** - just plain text
- ❌ **Manual extraction** - you write all parsing logic
- ❌ **Not AI-optimized** - requires post-processing
- ❌ **No structured data** - need to extract everything manually
- ❌ **~40+ lines** of code for basic functionality


```python
def scrape_with_beautifulsoup(url: str) -> Dict[str, Any]:
    """
    BeautifulSoup: Traditional web scraping
    Notice how much manual work is required!
    """
    start_time = time.time()

    try:
        # Step 1: Manual HTTP request
        response = requests.get(url, timeout=10)
        response.raise_for_status()

        # Step 2: Parse HTML manually
        soup = BeautifulSoup(response.content, "html.parser")

        # Step 3: Extract title manually
        title = soup.find("title").text if soup.find("title") else "No title"

        # Step 4: Remove script and style tags manually
        for script in soup(["script", "style"]):
            script.decompose()

        # Step 5: Extract text - NOT formatted for AI
        text = soup.get_text()

        # Step 6: Manual cleanup (still not great)
        lines = (line.strip() for line in text.splitlines())
        chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
        text = " ".join(chunk for chunk in chunks if chunk)

        duration = time.time() - start_time

        return {
            "tool": "BeautifulSoup",
            "title": title,
            "content_preview": text[:300] + "...",
            "has_markdown": False,  # ❌ NO MARKDOWN
            "has_links": False,  # ❌ Need manual extraction
            "has_media": False,  # ❌ Need manual extraction
            "ai_ready": False,  # ❌ Not optimized for AI
            "duration": duration,
            "code_complexity": "HIGH - ~40+ lines",
        }
    except Exception as e:
        return {"tool": "BeautifulSoup", "error": str(e)}


# Test BeautifulSoup
print("Testing BeautifulSoup...")
bs_result = scrape_with_beautifulsoup(TEST_URL)
print(f"\n✓ Scraped in {bs_result['duration']:.2f}s")
print(f"✗ Markdown support: {bs_result['has_markdown']}")
print(f"✗ AI-ready: {bs_result['ai_ready']}")
print(f"\nContent preview:\n{bs_result['content_preview']}")
```

## 3. Crawl4AI - The AI-First Game Changer

### Why Crawl4AI Dominates:

✅ **Markdown Output** - Perfect for LLMs and AI models  
✅ **4 Lines of Code** - vs 40+ (BS) or 100+ (Scrapy)  
✅ **AI-First Design** - Built specifically for AI applications  
✅ **Automatic Extraction** - Links, media, metadata all handled  
✅ **Clean Formatting** - Pre-processed for AI consumption  
✅ **Async Support** - High performance  
✅ **Zero Configuration** - Works out of the box

### The Magic of Crawl4AI:


```python
async def scrape_with_crawl4ai(url: str) -> Dict[str, Any]:
    """
    Crawl4AI: Watch how simple this is!
    4 lines vs 40+ with BeautifulSoup
    """
    start_time = time.time()

    # That's it! This is ALL you need:
    async with AsyncWebCrawler(verbose=False) as crawler:
        result = await crawler.arun(url=url)

    duration = time.time() - start_time

    return {
        "tool": "Crawl4AI",
        "title": result.metadata.get("title", "No title"),
        "markdown_preview": result.markdown[:300] + "...",
        "has_markdown": True,  # ✅ FULL MARKDOWN!
        "has_links": True,  # ✅ AUTO EXTRACTED
        "has_media": True,  # ✅ AUTO EXTRACTED
        "ai_ready": True,  # ✅ PERFECT FOR AI
        "duration": duration,
        "code_complexity": "LOW - ~4 lines!",
        "links_count": {
            "internal": len(result.links.get("internal", [])),
            "external": len(result.links.get("external", [])),
        },
        "media_count": {
            "images": len(result.media.get("images", [])),
            "videos": len(result.media.get("videos", [])),
            "audios": len(result.media.get("audios", [])),
        },
        "success": result.success,
    }


# Test Crawl4AI
print("Testing Crawl4AI...")
crawl4ai_result = await scrape_with_crawl4ai(TEST_URL)
print(f"\n✅ Scraped in {crawl4ai_result['duration']:.2f}s")
print(f"✅ Markdown support: {crawl4ai_result['has_markdown']}")
print(f"✅ AI-ready: {crawl4ai_result['ai_ready']}")
print(f"✅ Links extracted: {crawl4ai_result['links_count']}")
print(f"✅ Media extracted: {crawl4ai_result['media_count']}")
print(f"\nMarkdown preview:\n{crawl4ai_result['markdown_preview']}")
```
